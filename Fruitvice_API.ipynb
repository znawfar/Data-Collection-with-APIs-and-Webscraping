```python
import requests
import json
import pandas as pd
```

```python
#Obtain the fruitvice API data using requests.get("url") function.
#The data is in a json format.
data = requests.get("https://www.fruityvice.com/api/fruit/all")
```

```python
#Retrieve results using json.loads() function.
results = json.loads(data.text)
```

```python
#Convert our json data into pandas data frame.
pd.DataFrame(results)
```

>The result is in a nested json format.

> The **'nutrition'** column contains **multiple subcolumns**. The data requires to be **normalized** (flattened).

```python
df = pd.json_normalize(results)
```

```python
df
```
